```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salario = np.array([100, 200, 300, 500, 150])
salario
```

```python
plt.plot(salario, label="teste") 
```

```python
plt.plot(salario, c='black')
```

```python
plt.plot(salario, c='Black',ls ='--')
```

```python
plt.plot(salario, ls='--', marker='s')
```

```python
plt.plot(salario, c='black', ls='--',marker='s', ms=18)
plt.show()
```

```python
salarios_marcos = np.array([100,200,400,300])
salarios_gileno = np.array([50,300,500,450])
```

```python
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label= 'Salarios Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='s', ms=8, label= 'Salarios Gileno')
plt.show()
```

```python
#colocando legenda

plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label= 'Salarios Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='s', ms=8, label= 'Salarios Gileno')
plt.legend()
plt.show()
```

```python
#colocando legenda

plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label= 'Salarios Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8, label= 'Salarios Gileno')
plt.legend()
plt.show()
```

```python
#colocando legenda

plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label= 'Salarios Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='s', ms=8, label= 'Salarios Gileno')
plt.legend(loc='upper right')
plt.show()
```

```python
#colocando mais uma linha

salarios_pedro = [200,150,300,700]
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=10, label= 'Salarios Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='s', ms=10, label= 'Salarios Gileno')
plt.plot(salarios_pedro, c='green', ls='--', marker='o', ms=10, label= 'Salarios Pedro')
plt.legend(loc='upper left')
plt.show()
```
